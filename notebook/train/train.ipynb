```python
df['matchDate']=pd.to_datetime(df['matchDate'])
```

```python
df.dtypes
```

```python
df['day_code']=df['matchDate'].dt.dayofweek
```

```python
df
```

```python
df['target']=(df['teamoneScore']-df['teamtwoScore'])
```

```python
df=df[df['target']!=0]
```

```python
df.info()
```

```python
df=df.dropna()
```

```python
df.info()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
predictors=['teamoneID','teamtwoID','day_code','seasonID']
```

```python
rf=RandomForestClassifier(n_estimators=50,min_samples_split=30,random_state=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df[predictors].values
y=df['target'].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1,stratify=y)
```

```python
X_train
```

```python
rf.fit(X_train,y_train)
```

```python
preds=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(y_test,preds)
```

```python
acc
```
