```python
import os
import pandas as pd
```

```python
scr = pd.read_csv('/content/scheduler-2.csv')
```

```python
team = pd.read_csv('/content/team_esport (1).csv')
```

```python
scr.head(3)
```

```python
tmp1=team['shortNameTeam']
nameDict={}

for index,i in enumerate(tmp1):
  if i not in nameDict.keys() :
    nameDict[i]=int(index+1)
  print(index)
```

```python
print(nameDict)
```

```python
nameDict['PVB']
```

```python
# Iterate through the DataFrame and update values with error handling
for index in scr.index:
    try:
        scr.loc[index, 'teamoneID'] = nameDict[scr.loc[index, 'teamoneID']]
    except KeyError:
        print(scr.loc[index, 'teamoneID'])
```

```python
scr['teamoneID']
```

```python
# Iterate through the DataFrame and update values with error handling
for index in scr.index:
    try:
        scr.loc[index, 'teamtwoID'] =nameDict[scr.loc[index, 'teamtwoID']]
    except KeyError:
        print(scr.loc[index, 'teamtwoID'])
```

```python
scr.head(5)
```

```python
# Iterate through the DataFrame and update values with error handling
for index in scr.index:
    try:
        scr.loc[index, 'teamoneScore'] =scr.loc[index, 'teamoneScore'] if scr.loc[index, 'teamoneScore'] not in ['W','FF'] else ''
    except KeyError:
        scr.loc[index, 'teamoneScore'] = None  # or another placeholder like -1
```

```python
# Iterate through the DataFrame and update values with error handling
for index in scr.index:
    try:
        scr.loc[index, 'teamtwoScore'] =scr.loc[index, 'teamtwoScore'] if scr.loc[index, 'teamtwoScore'] not in ['W','FF'] else ''
    except KeyError:
        scr.loc[index, 'teamtwoScore'] = None  # or another placeholder like -1
```

```python
scr.to_csv('scheduler_preprocessed.csv')
```
