```python
def get_element_by_index(s, index):
  """
  Lấy phần tử trong set bằng chỉ số mà không xóa.

  Args:
    s: Set cần lấy phần tử.
    index: Chỉ số phần tử muốn lấy.

  Returns:
    Giá trị phần tử nếu tìm thấy, None nếu không tìm thấy.
  """
  for i, item in enumerate(s):
    if i == index:
      return item
  return None
```

```python
matchDate=[]
seasonID=[]
matchtypeID=[]
teamoneID=[]
teamtwoID=[]
teamoneScore=[]
teamtwoScore=[]
stadiumID=[]
check=[    '/wiki/VCS/2024_Season/Spring_Season',
    '/wiki/VCS/2024_Season/Summer_Season',    '/wiki/VCS/2023_Season/Summer_Season' ,   '/wiki/VCS/2023_Season/Spring_Season']
from urllib.error import HTTPError
for index,season in enumerate(new_list2):
  try:
    html = urlopen(f'https://lol.fandom.com{season}')
  except HTTPError as e:
    continue
  bs = BeautifulSoup(html, 'html.parser')
  r=bs.find_all('div',{'id':'matchlist-content-wrapper'})
  for q in r:
    matchlist = q.find_all('tr', {'data-date':True})
    for h in matchlist:
      matchDate.append(h.attrs['data-date'])
      seasonID.append(index+1)
      matchtypeID.append(1)
      if season in check:
        stadiumID.append(2)
      else:
        stadiumID.append(1)
  f=bs.find_all('span',{'class':'teamimage-left'})
  for g in f:
   teamoneID.append(g.find_previous_siblings()[0].get_text())
  r=bs.find_all('div',{'class':'matchlist-tab-wrapper'})
  for q in r:
    h=q.find_all('span',{'class':'teamimage-right'})
    for o in h:
      n=o.findNextSibling()
      if not n.find('a'):
        teamtwoID.append(n.get_text())
  u=bs.find_all('td',{'class':'matchlist-score'})
  for index2,x in enumerate(u):
    if (index2+1)%2==0:
      teamtwoScore.append(x.get_text())
    else:
      teamoneScore.append(x.get_text())
```
